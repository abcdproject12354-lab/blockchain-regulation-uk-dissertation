```python
import pandas as pd
import re
```

```python
hansard = pd.read_excel("/data/FCA analysis.xlsx")
fca = pd.read_excel("/data/Hansard.xlsx")

print("Hansard rows:", hansard.shape)
print("FCA rows:", fca.shape)
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
hansard["clean_text"] = hansard["Section"].apply(clean_text)
fca["clean_text"] = fca["Speaker/Section"].apply(clean_text)
```

```python
hansard.to_csv("/data/hansard_clean.csv", index=False)
fca.to_csv("/data/fca_clean.csv", index=False)

print("Cleaning complete")
```
