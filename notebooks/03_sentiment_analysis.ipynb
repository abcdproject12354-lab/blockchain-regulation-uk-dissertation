```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
hansard = pd.read_csv("/content/hansard_clean.csv")
fca = pd.read_csv("/content/fca_clean.csv")
```

```python
def get_sentiment(text):
    return TextBlob(str(text)).sentiment.polarity
```

```python
hansard["sentiment"] = hansard["clean_text"].apply(get_sentiment)
fca["sentiment"] = fca["clean_text"].apply(get_sentiment)
```

```python
print("Hansard average sentiment:", hansard["sentiment"].mean())
print("FCA average sentiment:", fca["sentiment"].mean())
```

```python
plt.figure()
plt.boxplot([hansard["sentiment"], fca["sentiment"]], labels=["Hansard", "FCA"])
plt.title("Sentiment Comparison")
plt.show()
```
