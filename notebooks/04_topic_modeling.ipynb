```python
import pandas as pd
hansard = pd.read_csv("/content/hansard_clean.csv")
fca = pd.read_csv("/content/fca_clean.csv")
```

```python
h_docs = [text.split() for text in hansard["clean_text"]]
f_docs = [text.split() for text in fca["clean_text"]]
```

```python
import gensim.corpora as corpora
h_dict = corpora.Dictionary(h_docs)
f_dict = corpora.Dictionary(f_docs)

h_corpus = [h_dict.doc2bow(doc) for doc in h_docs]
f_corpus = [f_dict.doc2bow(doc) for doc in f_docs]
```

```python
import gensim
h_lda = gensim.models.LdaModel(h_corpus, num_topics=3, id2word=h_dict, passes=10)
f_lda = gensim.models.LdaModel(f_corpus, num_topics=3, id2word=f_dict, passes=10)
```

```python
print("Hansard Topics")
for t in h_lda.print_topics():
    print(t)

print("\nFCA Topics")
for t in f_lda.print_topics():
    print(t)
```
