```python
hansard = pd.read_csv("/content/hansard_clean.csv")
fca = pd.read_csv("/content/fca_clean.csv")
```

```python
def tokenize(text):
    return text.split()

h_tokens = " ".join(hansard["clean_text"]).split()
f_tokens = " ".join(fca["clean_text"]).split()
```

```python
h_freq = Counter(h_tokens).most_common(20)
f_freq = Counter(f_tokens).most_common(20)

pd.DataFrame(h_freq, columns=["word","count"])
```

```python
def plot_freq(freq, title):
    words = [w for w,_ in freq]
    counts = [c for _,c in freq]

    plt.figure()
    plt.bar(words, counts)
    plt.xticks(rotation=90)
    plt.title(title)
    plt.show()

plot_freq(h_freq, "Hansard Keyword Frequency")
plot_freq(f_freq, "FCA Keyword Frequency")
```
